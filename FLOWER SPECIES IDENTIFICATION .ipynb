```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# load dataset into Pandas DataFrame

data = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
#DATA STANDARDIZATION SECTION
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = ["sepal length","sepal width","petal length","petal width"]
# FEATURES SEPERATION
X = data.loc[:,features].values
y = data.loc[:,['target']].values
```

```python
X
```

```python
# standardizing the features 
d = StandardScaler().fit_transform(X)
```

```python
d
```

```python
#  PRINCIPLE COMPONENT ANALYSIS PROJECTION TO 2D
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
principal_components = pca.fit_transform(X)
new_data = pd.DataFrame(data = principal_components , columns = ["principal1","principal2"])
```

```python
principal_components
```

```python
data[["target"]] # double brackets are used to get the DataFrame 
```

```python
combined_df = pd.concat([new_data,data[["target"]]],axis = 1)
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (10,10)) # it creates a new figure
image = fig.add_subplot(1,1,1) #  dimensions of the figure 
image.set_xlabel("PRINCIPAL COMPONENT 1",fontsize = 20)
image.set_ylabel("PRINCIPAL COMPONENT 2",fontsize = 20)
image.set_title(" TWO COMPONENTS PCA ",fontsize = 30)
targets = ["Iris-setosa","Iris-versicolor",'Iris-virginica']
colors = ["r","g","b"]
for target,color in zip(targets,colors):
    indicesToKeep = combined_df['target'] == target
    image.scatter(combined_df.loc[indicesToKeep,"principal1"],combined_df.loc[indicesToKeep,"principal2"],c =color,s = 50)
image.legend(targets)
image.grid()
```

```python
# PCA To Speed Up Machine Learning algorithms
# Download and load the data
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
# test_size: what proportion of original data is used for test set
train_img, test_img, train_lbl, test_lbl = train_test_split( mnist.data, mnist.target,test_size=1/7.0,random_state=0)

```

```python
mnist.data
```

```python
mnist.target
```

```python
# DATA STANDARDIZATION 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# fit on training set only
scaler.fit(train_img)
```

```python
# APPLY TRANSFORM TO BOTH TRAIN AND TEST DATA 
```

```python
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
```

```python
from sklearn.decomposition import PCA
pca = PCA(.95)
```

```python
pca.fit(train_img)
```

```python
# APPLY THE MAPPING TO BOTH THE TRAINING SET AND TESTING DATA SET
```

```python
train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
# APPLIED LOGISTIC REGRESSION TO THE TRANSFORMED DATA 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# MAKE AN INSTANCE OF THE MODEL 
# ALL PARAMETERS WHICH ARE NOT SPECIFIED ARE SET TO THEIR DEFAULTS
# DEFAULT SOLVER IS INCREDIBLE SLOW WHICH IS WHY IT WAS CHANGED TO 'lbfgs'
lr = LogisticRegression(solver = "saga")
```

```python
# MODEL IS LEARNING THE RELATIOIN SHIP BETWEEN DIGITS AND LABELS
```

```python
lr.fit(train_img,train_lbl)
```

```python
lr.predict(test_img[0].reshape(1,-1))
```

```python
# THE FOLLOWING CODE IS THE PREDICTION FOR MULTIPLE VARIABLES AT A TIME 
# PREDICT FOR ONE OBSERVATION [ IMAGE]
lr.predict(test_img[0:10])
```

```python
# PERFORMANCE OF THE MODEL 

```

```python
lr.score(test_img,test_lbl)
```
